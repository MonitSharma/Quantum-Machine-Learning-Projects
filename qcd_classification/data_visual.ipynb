## PCA


```python
features_label = ['dndy','v2','v3','v4', 'v5',	'psi2',
	'psi3',
     'psi4',
    	'psi5',
    'ptspec_bin0',
     'ptspec_bin',
     'ptspec_bin2',
    	'ptspec_bin3',
     'ptspec_bin4',
    	'ptspec_bin5',
    	'ptspec_bin6',
    	'ptspec_bin7',
    	'ptspec_bin8']



fetures=dataset[features_label]
scalar=StandardScaler()
scalar.fit(fetures)
scale_data=scalar.transform(fetures)

new_futures=['PC-1','PC-2','PC-3','PC-4','PC-5','PC-6']
pca=PCA(n_components=6)
pca.fit(scale_data)
X_pca=pca.transform(scale_data)
X_pca=pd.DataFrame(X_pca, columns=new_futures)
print(X_pca.shape)
X_pca.head(5)
```

```python
nr_row=1000
color=['r' if dataset['eos_type'][i]==1  else 'b' for i in range(nr_row)]
future_1=X_pca['PC-1'][:nr_row]
future_2=X_pca['PC-2'][:nr_row]
future_3=X_pca['PC-3'][:nr_row]
future_4=X_pca['PC-4'][:nr_row]
future_5=X_pca['PC-5'][:nr_row]
future_6=X_pca['PC-6'][:nr_row]


```

```python
fig1=plt.figure()



ax1=Axes3D(fig1)
#ax1 = fig1.gca(projection='3d')
ax1.scatter(future_1,future_2,future_3,color=color)
ax1.set_xlabel('PC-1')
ax1.set_ylabel('PC-2')
ax1.set_zlabel('PC-3')



plt.show()
```

```python
fig2=plt.figure()

ax2=Axes3D(fig2)
ax2.scatter(future_4,future_5,future_6,color=color)
ax2.set_xlabel('PC-4')
ax2.set_ylabel('PC-5')
ax2.set_zlabel('PC-6')

plt.show()
```
